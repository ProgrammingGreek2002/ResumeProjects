### First Visualization:

In this graph you can see a steady increase of COVID-19 from the year 2020 to 2022. At first glance it seems as though the rate of COVID-19 cases are steadily increasing, but when you take all information into account it becomes apparent that vaccines have greatly helped society. In the years 2020 and 2021 vaccines weren't readily available, allowing the virus to run through society without much issue. Only near the middle/end of 2022 did they become easy to access allowing the majority of the population to be protected. That is why even though we are a quarter of the way through 2023 the amount of people sick this year is miniscule compared to last year. The vaccine allowed for this to occur,  showing that it is in fact effective in combatting the disease.

The data can be accessed from the link provided below:

https://open.canada.ca/data/en/dataset/f4f86e54-872d-43f8-8a86-3892fd3cb5e6

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://data.ontario.ca/dataset/f4f86e54-872d-43f8-8a86-3892fd3cb5e6/resource/8a88fe6d-d8fb-41a3-9d04-f0550a44999f/download/daily_change_in_cases_by_phu.csv')

# Convert the date column to a datetime object
df['Date'] = pd.to_datetime(df['Date'])

# Extract the year from the date and add it to a new column
df['Year'] = df['Date'].dt.year

# Group the data by year and sum up the total number of cases for each year
cases_by_year = df.groupby('Year')['Total'].sum()

# Create a bar graph of the total number of cases per year
cases_by_year.plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Total Number of Cases')
plt.title('Total Number of COVID-19 Cases by Year')
plt.show()
```

### Second Visualization:

The visualization displayed in the graph below show the people that are hospitalized with COVID-19, and the ones who are in need of ICU among the hospitalized. The first dose of vaccinations started on December 15, 2020, followed by the second dose on May 31, 2021. This is significant because the first two peaks in the graph coincide with the period when the vaccines were given out. After the administration of the vaccines, a visible decline in the numbers of hospitalized and ICU patients can be seen in the visualization. There is a third peak in the graph however that happens right after New Year's since many people were violating the COVID-19 restrictions. ```python
import pandas as pd
import matplotlib.pyplot as plt

#import and extract dataset
df = pd.read_csv('https://data.ontario.ca/dataset/8f3a449b-bde5-4631-ada6-8bd94dbc7d15/resource/e760480e-1f95-4634-a923-98161cfb02fa/download/region_hospital_icu_covid_data.csv', parse_dates=['date'], index_col=['date'])
month = df.resample('M').mean()

#plot
plt.plot(month.index, month['hospitalizations'], label='Covid Hospitalizations')
plt.plot(month.index, month['icu_current_covid'], label='COVID ICU Patients')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Number of Patients')
plt.title('COVID-19 Hospitalizations and COVID-19 ICU Patients')
plt.xticks(rotation=45)
plt.show()
```

### Third Visualization:

In this graph it can be seen that there are high and low points at certain intervals. Although, not readily apparent  unless you take a closer look and correlate the dates. ```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://data.ontario.ca/dataset/c43fd28d-3288-4ad2-87f1-a95abac706b8/resource/3273c977-416f-407e-86d2-1e45a7261e7b/download/deaths_fatality_type.csv')
# print(df.head(300))

df['date'] = pd.to_datetime(df['date'])

# group data by date and sum death_covid column
df_covid = df.groupby('date')['death_covid'].sum().reset_index()

# plot death_covid column
plt.plot(df_covid['date'], df_covid['death_covid'])
plt.xlabel('Date')
plt.ylabel('Deaths due to COVID-19')
plt.title('Deaths due to COVID-19 in Ontario over time')
plt.xticks(rotation=45)
```

### Fourth Visualization:

For the last visualization, it goes more in depth at the total COVID-19 related deaths. The four lines in the graph are seperated as Total number of deaths involving COVID-19, Number of deaths with “COVID-19 as the underlying cause of death”, Number of deaths with “COVID-19 contributed but not underlying cause” and Number of deaths where the “Cause of death unknown”. As you can see, COVID-19 has worsened the health of people that were already suffering from other sicknesses. This can be seen in the graph where COVID-19 contributed in death but were not the underlying cause and the unknown death causes. They both react similarly towards how COVID-19 as the underlying cause line did, suggesting a correlation between vaccination and saving lives of those who may have otherwise died due to COVID-19 contributions.

The data can be accessed from the link provided below:

https://data.ontario.ca/en/dataset/deaths-involving-covid-19-by-fatality-type

```python
import pandas as pd
import matplotlib.pyplot as plt

#import dataset
data = pd.read_csv('https://data.ontario.ca/dataset/c43fd28d-3288-4ad2-87f1-a95abac706b8/resource/3273c977-416f-407e-86d2-1e45a7261e7b/download/deaths_fatality_type.csv', parse_dates=['date'], index_col='date')

#extract data
monthly_data = data.resample('M').sum()
monthly_data.plot(kind='line', y=['deaths_total', 'death_covid', 'death_covid_contrib', 'death_unknown_missing'], figsize=(10,5), \
                  label=['Total Deaths', 'COVID-19 as Underlying Cause', 'COVID-19 Contributing Deaths', 'Unknown Deaths'])

#plot
plt.xlabel('Date')
plt.ylabel('Number of Deaths')
plt.title('Deaths relating to COVID-19')
plt.legend(loc='upper right')
plt.show()
```
